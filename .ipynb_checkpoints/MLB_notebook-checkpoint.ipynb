## Baseball Analytics, A brief history

Baseball is America's oldest major sport, in 1856 it was referred to as "America's Pastime"(Tygiel, 2000) and by 1869 the Cincinatti Red Stockings became the United States first professional sports team. As Baseball was America's first major sport, it was also it's first sport to be subjected to rigoroous statsitical analysis. In 1971 the Society for American Baseball research (SABR) was founded, who's acronym gave rise to the Sabermetrics, which broadly refers to the use of statitical analysis and prognostication to better understand player performance.

## Why do we want to better understand baseball?

- There are three people who could benefit form a more thorough understanding of baseball. In no particular order they are:

    - The Fan
    - The Owner/Scout
    - And of course, the Gambler
 
### The Use Case for the Fan:

"Yes he was good, but *he* was better".

Variations of this argument have taken place in the nations bars for as long as there has been the sport. Stats allow us to track how well a player performed and gives us objective metrics for comparison sake.

### The use case for the manager:

Moneyball, the identifaction of undervalued players. Historically, certain statistcs were valued as more important than others. Billy Beane of the A's was famous for identifying stats such as *walks*, that while they didn't receive the same prestige as the more exciting *hits*, were just as impactful on winning. He used this types of analysis to find players who were more impactful than the popular metrics indicated.

## Importing Data

##### For this project we will be utlizing the pybaseball library. 

This is a very nifty tool which allows us to scrape publicly available baseball statistics and easily export them in tabular format.

```python
# Import necessary packages from bybaseball library
from pybaseball import batting_stats_bref, batting_stats, pitching_stats
```

```python
# grabs batting data for the 2020-2024 seasons for batters with >200 plate appearances.

batting_20_23 = batting_stats(2020,2023,qual = 200)
```

```python
# converts each season to a dataframe
batting_df_raw = pd.DataFrame(batting_20_23)
```

```python
# Trims down the batting dataframe to classic counting stats and a few advanced metrics
batting_df= batting_df_raw.loc[::,'IDfg':'WAR']
```

```python
# Imports pitching data for the 2020-2024 season for pitchers with >50 innings pitcheed
pitching_20_23 = pitching_stats(2020,2023,qual = 50)
```

```python
# converts each season to a data frame
pitching_df_raw = pd.DataFrame(pitching_20_23)
```

```python
# Trims down pitching dataframe to classic counting stats and a few advanced metrics
pitching_df = pitching_df_raw.loc[::,'IDfg':'BUH%']
```

## Exploration

##### One metric that has become very popular is WAR, or "wins above replacement". A player is given a WAR rating which is meant to indicate the number of wins they would contribute to their team compared to a 'replacement level player'. A replacement level player is a player that could be added by any team at any time for minimal cost, and is considered esentially the baseline for servicability in the major leagues.

```python
# Counts the number of players at or above 0 war and below 0 war
above_0_war = len(batting_df[batting_df['WAR']>=0])
below_0_war = len(batting_df[batting_df['WAR']<0])
```

```python
fig1,ax1 = plt.subplots()
labels = ['Above Replacement Level','Below Replacement Level']
sizes = [above_0_war,below_0_war]
ax1.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.show()
```

#### We can see on our ever controversial pie chart that ~ 15% of players are considered below replacement level

### What do we see?

#### - Above we see that the distribution of WAR is right skewed with no outliers to the bottom and some to the top. This tracks with the intuition of the WAR metric, if a player is performing poorly enough to generate a WAR nearing -2.0 they are likely to be replaced by a 'replacement player'

#### - Also we see that the range for pitching WAR is narrower with a lower median and mode. This is easily explained by the fact that pitchers simply play less often than hitters, giving them fewer opportunities to contribute to winning and accrue this stat.
